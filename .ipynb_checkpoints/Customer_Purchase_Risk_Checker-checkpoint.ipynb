```python
email = "user123@gmail.com"
umur = 16
harga = 1500000
status_pembayaran = False
jenis_orderan = "Elektronik"
```

```python
ada_at = "@" in email
```

```python
if ada_at:
    bagian_email = email.split("@")
    username = bagian_email[0]
    domain = bagian_email[1]
else:
    username = ""
    domain = ""

teks_sebelum_at = username != ""
teks_setelah_at = domain != ""
domain_punya_titik = "." in domain

email_valid = ada_at and teks_sebelum_at and teks_setelah_at and domain_punya_titik
```

```python
risk_score = 0

if not email_valid:
    risk_score = risk_score + 2

if umur < 18:
    risk_score = risk_score + 1

if harga > 1000000:
    risk_score = risk_score + 1

if status_pembayaran == False:
    risk_score = risk_score + 2
```

```python
if risk_score >= 0 and risk_score <= 2:
    risk_label = "SAFE"
elif risk_score >= 3 and risk_score <= 4:
    risk_label = "SUSPICIOUS"
else:
    risk_label = "HIGH RISK"
```

```python
print("=== CUSTOMER PURCHASE RISK CHECKER ===")
print("Email                :", email)
print("Umur                 :", umur)
print("Harga                :", harga)
print("Status Pembayaran    :", status_pembayaran)
print("Jenis Orderan        :", jenis_orderan)

print("\n=== EMAIL VALIDATION ===")
print("Ada @                :", ada_at)
print("Teks sebelum @       :", teks_sebelum_at)
print("Teks setelah @       :", teks_setelah_at)
print("Domain punya titik   :", domain_punya_titik)
print("Email valid          :", email_valid)

print("\n=== RISK SCORING DETAIL ===")
print("Email invalid        :", not email_valid)
print("Umur < 18            :", umur < 18)
print("Harga > 1 juta       :", harga > 1000000)
print("Belum dibayar        :", status_pembayaran == False)

print("\n=== FINAL RESULT ===")
print("Risk Score           :", risk_score)
print("Risk Label           :", risk_label)
```
